```python
import numpy as np
import matplotlib.pyplot as plt

def d(a, b):
    return np.sqrt(np.sum((np.array(a)-np.array(b))**2))

a = [0, 0]
b = [3, 1]
c = [3, 4]

print("check?", d(a, b) + d(b, c), ">", d(a, c))
print("check?", d(b, c) + d(c, a), ">", d(b, a))
print("check?", d(b, a) + d(a, c), ">", d(b, c))

# 点をプロット
plt.plot([a[0], b[0], c[0]], [a[1], b[1], c[1]], 'bo')

# 線分をプロット
plt.plot([a[0], b[0]], [a[1], b[1]], 'r')
plt.plot([b[0], c[0]], [b[1], c[1]], 'r')
plt.plot([c[0], a[0]], [c[1], a[1]], 'r')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-x**2)

def g(x):
    return np.exp(-(x-2)**2/4)

def h(x):
    return np.exp(-(x-4)**2/2)

dx = 0.1

def d(a, b):
    x = np.arange(-10, 10, dx)
    return sum(np.abs(a(x)-b(x))*dx)

print("check?", d(f, g) + d(g, h), ">", d(f, h))
print("check?", d(g, f) + d(f, h), ">", d(g, h))
print("check?", d(f, h) + d(h, g), ">", d(f, g))

x = np.arange(-10, 10, dx)

# f(x), g(x), h(x)のグラフをプロット
plt.plot(x, f(x), label='f(x)')
plt.plot(x, g(x), label='g(x)')
plt.plot(x, h(x), label='h(x)')

plt.legend()
plt.show()
```

```python
def synthetic_sensor_data(num_sensors, num_examples):
    mean = np.array([(i+1)*15 for i in range(num_sensors)])
    cov = np.ones((num_sensors, num_sensors))*0.5
    cov += np.identity(num_sensors)*0.5
    X = np.random.multivariate_normal(mean,cov, num_examples)
    interval = 800
    X[interval:,1] += np.random.normal(loc=2, scale=1, size=len(X)-interval)
    return X
data = synthetic_sensor_data(3, 1000)
data
```
