```python
import numpy as np
import torch
 
#入れ子のlistを渡して作成
t = torch.tensor([[1, 2], [3, 4.]])

#0~9までの数値で初期化された1次元のTensor
a = torch.arange(0, 10)

#正規分布で100×10のTensorを作成
c = torch.randn(100, 10)

print(c.size())
print(c)

print(a.size())
print(a)

print(t.size())
print(t)
```

```python
#numpyメソッドを使用して，ndarrayに変換
a = torch.tensor([[1, 2], [3, 4]])
x = a.numpy()

print(a)
print(x)
```

```python
t = torch.tensor([[1, 2, 3], [4, 5, 6]])

#スカラーの添字で指定
print(t[0, 2])

#スライスで指定
print(t[:, :2])

#添字のリストで指定
print(t[:, [1,2]])

#マスク配列を使用して，3より大きい部分のみを選択
print([t>3])

#[0, 1]要素を100にする
t[0,1] = 100

print(t)

#スライスを使用した一括代入
t[:, 1] = 200

print(t)

#マスク配列を使用して，特定条件の要素のみ置換
t[t > 10] = 20
print(t)
```

```python
#100×10のテストデータを用意
X = torch.randn(100, 10)

y = X * 2 + torch.abs(X)
print(y)

#平均値を求める
m = torch.mean(X)

#関数ではなく，メソッドとしても利用できる
m = X.mean()

print(m)

#集計結果を0次元のTensorでitemメソッドを使用して，
#値を取り出すことができる
m_value = m.item()
print(m_value)

#集計して列ごとに平均値を計算している
m2 = X.mean(0)
print(m2)
```

```python
x1 = torch.tensor([[1, 2], [3, 4.]])
x2 = torch.tensor([[10, 20, 30], [40, 50, 60.]])

print(x2.size())

print(x1.view(4, 1))
print(x1)

print(x1.reshape(4, 1))
print(x1)

print(x2)
#2×3を転置して3×2にする
print(x2.t())
print(x2)

#dim=1に対して結合することで2×5のTensorを作る
torch.cat([x1, x2], dim=1)

#64×32×3のデータが100個
hwc_img_data = torch.rand(100, 64, 32, 3)
print(hwc_img_data)
print(hwc_img_data.size())
#64と32を入れ替えたのちに，32と3を入れ替える
chw_img_data = hwc_img_data.transpose(1, 2).transpose(1, 3)
print(chw_img_data)
print(chw_img_data.size())
```

```python
m = torch.randn(100, 10)
v = torch.randn(10)

#内積
d = torch.dot(v, v)
print(d)

#100×10の行列と長さ10のベクトルの積
#結果は長さ100のベクトル
v2 = torch.mv(m, v)
print(v2)

#行列積
m2 = torch.mm(m.t(), m)
print(m2)

#特異値分解
u, s, v = torch.svd(m)
print(u, s, v)
```

```python
x = torch.randn(100, 3)
#微分の変数として扱う場合はrequires_gradフラグをTrueにする
a = torch.tensor([1, 2, 3.], requires_grad=True)
print(x)
print(a)

#計算をすることで，自動的に計算グラフが構築されていく
y = torch.mv(x, a)
o = y.sum()
print(y)
print(o)

#微分を実行する
o.backward()
```
